### <b>Bar chart For Categorical Columns

```python
plt.bar(gender_counts.index,gender_counts.values,color="blue",edgecolor="black")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution of Cancer Patients")
plt.show()
```

```python
countries_count=df["Country_Region"].value_counts()
```

```python
plt.bar(countries_count.index,countries_count.values)
plt.xlabel("Countries")
plt.ylabel("Count")
plt.title("Distribution of cancer patients across Different countries")
plt.show()
```

```python
cancer_stage_counts=df["Cancer_Stage"].value_counts()
plt.bar(cancer_stage_counts.index,cancer_stage_counts.values)
plt.xlabel("Cancer Stages")
plt.ylabel("Patients Counts")
plt.title("Distribution of patients across Cancer Stages")
plt.show()
```

### <b>Histogram For Numerical Columns

```python
plt.hist(df["Age"],bins=20,edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Cancer Patients")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.hist(df["Alcohol_Use"],bins=15,edgecolor="black")
plt.xlabel("Alcohol Usage")
plt.ylabel("Count")
plt.title("Alcohol Usage Distribution of Cancer Patients")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Load the dataset
df = pd.read_csv("global_cancer_patients_2015_2024.csv")

# Extract Age data
age_data = df["Age"]

# Create the histogram
plt.figure(figsize=(8, 6))
count, bins, ignored = plt.hist(age_data, bins=20, color='skyblue', edgecolor='black', alpha=0.6, density=True)

# Create and plot the KDE (Kernel Density Estimate) curve
kde = gaussian_kde(age_data)
x_vals = np.linspace(age_data.min(), age_data.max(), 500)
plt.plot(x_vals, kde(x_vals), color='darkblue', linewidth=2, label='Density Curve')

# Final plot settings
plt.title("Age Distribution with Density Curve")
plt.xlabel("Age")
plt.ylabel("Density")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
